```python
df['TotalCharges']  =  pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encode = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() == 2]

le = LabelEncoder()

for col in label_encode:
  df[col]= le.fit_transform(df[col])
```

```python
df.head(10)
```

```python
onehot = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique()>=2]

one_hot_df = pd.get_dummies(df[onehot]).astype(int)


df.drop(onehot, axis=1, inplace=True)
```

```python
df_com = pd.concat([df, one_hot_df], axis=1)
```

```python
df_com.head(5)
```

```python


X = df_com.drop('Churn', axis=1)
y = df_com['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state = 0)
```

```python
#scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

#Fit and transform on X_train, then transform only on x_test both

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```
