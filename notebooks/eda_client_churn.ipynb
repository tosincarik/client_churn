```python
df['TotalCharges']  =  pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Impute missing values (11 rows) with the median
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

label_encode = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() == 2]

le = LabelEncoder()

for col in label_encode:
  df[col]= le.fit_transform(df[col])
```

```python
df.head(10)
```

```python
onehot = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique()>=2]

one_hot_df = pd.get_dummies(df[onehot]).astype(int)


df.drop(onehot, axis=1, inplace=True)
```

```python
df_com = pd.concat([df, one_hot_df], axis=1)
```

```python
df_com.head(5)
```

```python


X = df_com.drop('Churn', axis=1)
y = df_com['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state = 0)
```

```python
#scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

#Fit and transform on X_train, then transform only on x_test both

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)
y_pred_log = log_reg.predict(X_test_scaled)
y_prob_log = log_reg.predict_proba(X_test_scaled)[:, 1]

print("Logistic Regression Report:")
print(classification_report(y_test, y_pred_log))
print("Logistic Regression ROC AUC:", roc_auc_score(y_test, y_prob_log))

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
y_prob_rf = rf.predict_proba(X_test_scaled)[:, 1]

print("\nRandom Forest Report:")
print(classification_report(y_test, y_pred_rf))
print("Random Forest ROC AUC:", roc_auc_score(y_test, y_prob_rf))
```

```python
def explain_churn(customer_data):
    messages = []

    # Use thresholds or categorical checks based on your domain knowledge and feature importance
    if customer_data.get('Contract_Month-to-month', 0) == 1:
        messages.append("Customer is on a month-to-month contract, which tends to increase churn risk.")

    if customer_data.get('InternetService_Fiber optic', 0) == 1:
        messages.append("Customer uses fiber optic internet service, which is associated with higher churn.")

    if customer_data.get('tenure', 0) < 12:
        messages.append("Customer has been with us less than a year, increasing churn risk.")

    if customer_data.get('MonthlyCharges', 0) > 70:
        messages.append("Customer's monthly charges are high, which may contribute to churn.")

    if customer_data.get('Contract_Two year', 0) == 1:
        messages.append("Customer has a two-year contract, which usually reduces churn risk.")

    if customer_data.get('TotalCharges', 0) < 1000:
        messages.append("Customer's total charges are relatively low, possibly indicating new or low-usage customers.")

    if customer_data.get('TechSupport_No', 0) == 1:
        messages.append("Customer does not have tech support, which can increase churn risk.")

    if customer_data.get('OnlineSecurity_No', 0) == 1:
        messages.append("Customer lacks online security services, which may contribute to churn.")

    if customer_data.get('PaymentMethod_Electronic check', 0) == 1:
        messages.append("Customer pays by electronic check, a payment method linked with higher churn.")

    if not messages:
        return "Customer shows low risk of churn based on key factors."

    return " ".join(messages)
```
